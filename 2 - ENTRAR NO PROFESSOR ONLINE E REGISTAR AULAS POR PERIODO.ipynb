```python
# BLOCO SEPARA DIA MES E ANO EM VARIAVEIS SEPARADAS;
# DEVE SER EXECULTADO 

dia = aulasdf['Data'].dt.day
mes = aulasdf['Data'].dt.month
ano = aulasdf['Data'].dt.year
print(dia, mes, ano)
```

```python
#registro de aulas  pela data
# ESSE BLOCO VC DEVE ESTA NA TELA 'Item do registro de aula'

i=0 # ESSE 'i' SERVIRÁ PARA CASO DE ERRO VC COMECAR O CADASTRO DO ULTIMO INDICE COM BASE NA TABELA DE REGISTRO DE AULAS. 

while i < len(aulasdf):
    try:
        datas = navegador.find_element(By.XPATH, '//*[@id="data"]')
        subconteudo = navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[2]/div/div/div/input')
        datas.clear()
        time.sleep(2)
        datas.send_keys(f'{dia[i]}/{mes[i]}/{ano[i]}')
        navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[1]/div[2]/div/label[1]/input').click()
        time.sleep(2)
        navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[1]/div/button').click()
        time.sleep(2)
        navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[1]/div/div/ul/li[3]/a').click() # mudar periodo aqui 
        time.sleep(2)
        navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[2]/div/button').click()
        time.sleep(1)
        subconteudo.send_keys(f'{aulasdf["Subconteudo"].values[i]}')
        pyautogui.press('enter')
        time.sleep(3)
        navegador.find_element(By.XPATH, '//*[@id="btn-save"]').click()
        time.sleep(1)
    except Exception as e:
        print(f"Erro encontrado: {str(e)}")
    
    i += 1
```
