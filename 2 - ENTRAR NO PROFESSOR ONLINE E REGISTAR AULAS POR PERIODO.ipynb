```python
# AUTOMAÇÃO SISTEMA PROFESSOR ONLINE
#ENTRAR NO PROFESSOR ONLINE COM SEU USUARIO E SENHA

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pyautogui


navegador = webdriver.Chrome()
#entrar no site do sice
navegador.get("https://professor.seduc.ce.gov.br/")

time.sleep(5)
pyautogui.hotkey('win', 'up') #MAXIMIZAR A TELA
time.sleep(5)
usuarioProf = "SEU USUARIO AQUI" #ADICIONE SEU USUARIO
senhaProf = "SUA SENHA AQUI" #ADICIONE SUA SENHA

#entrar na Conta 
navegador.find_element(By.XPATH, '//*[@id="top"]/div/div/div[2]/div/div/div/a').click()
navegador.find_element(By.XPATH, '//*[@id="username"]').send_keys(f"{usuarioProf}")
navegador.find_element(By.XPATH, '//*[@id="password"]').send_keys(f"{senhaProf}")
navegador.find_element(By.XPATH, '//*[@id="formLogin"]/div[3]/div/button').click()
```

```python
import pyautogui


# BLOCO DE REGITRO DE AULAS
# ESSE BLOCO ELE APENAS ENTRA NA TURMA E DICIPLINA PARA INICIAR OS PASSO DE INSERIR OS DADOS. VOCÊ PODE ENTRAR DE FORMA MANUAL


#VARIAVEIS UTILIZADAS VAI DEPENDER DO SEU CONTEUDO. UTILIZE ESSE VIDEO PARA ALTERAR OS XPHATH DE REFERENCIAS PARA SEU USO. 
# COMO COPIAR E SABER O XPATH AQUI -> https://youtu.be/poZ4COfSzfA

turmaInf = '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[2]/div/div/div/ul/li[3]/a/span[1]'
turmaDs = '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[2]/div/div/div/ul/li[2]/a/span[1]'

discArq = '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[3]/div/div/div/ul/li[2]/a/span[1]'
discPlan = '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[3]/div/div/div/ul/li[3]/a/span[1]'

navegador.find_element(By.XPATH, '//*[@id="navbar"]/ul/li[5]/a/div/div[2]').click()
time.sleep(1)
navegador.find_element(By.XPATH, '//*[@id="collapseSubMenuDiario"]/li[2]/a').click()
time.sleep(3)
navegador.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[1]/div/div/button').click()
navegador.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[1]/div/div/div/ul/li[2]/a/span[1]').click()
time.sleep(1)

navegador.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[2]/div/div/button').click()
navegador.find_element(By.XPATH, turmaInf).click()
time.sleep(1)

navegador.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[2]/div/div[1]/div/form/div/div[3]/div/div/button').click()
navegador.find_element(By.XPATH, discPlan).click()
time.sleep(1)

navegador.find_element(By.XPATH, '//*[@id="btn-add"]').click()
time.sleep(3)
```

```python
# BLOCO SEPARA DIA MES E ANO EM VARIAVEIS SEPARADAS;

dia = aulasdf['Data'].dt.day
mes = aulasdf['Data'].dt.month
ano = aulasdf['Data'].dt.year
print(dia, mes, ano)
```

```python
#registro de aulas  pela data
# ESSE BLOCO VC DEVE ESTA NA TELA 'Item do registro de aula'

i=0 # ESSE i SERVIRÁ PARA CASO DE ERRO VC COMECAR O CADASTRO DO ULTIMO INDICE COM BASE NA TABELA DE REGISTRO DE AULAS. 

while i < len(aulasdf):
    datas = navegador.find_element(By.XPATH, '//*[@id="data"]')
    subconteudo = navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[2]/div/div/div/input')
    datas.clear()
    time.sleep(2)
    datas.send_keys(f'{dia[i]}/{mes[i]}/{ano[i]}')
    navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[1]/div[2]/div/label[1]/input').click()
    time.sleep(2)
    navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[1]/div/button').click()
    time.sleep(2)
    navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[1]/div/div/ul/li[5]/a').click() # mudar periodo aqui 
    time.sleep(2)
    navegador.find_element(By.XPATH, '//*[@id="form-cadastro"]/div[2]/div[2]/div/button').click()
    time.sleep(1)
    subconteudo.send_keys(f'{aulasdf["Subconteudo"].values[i]}')
    pyautogui.press('enter')
    time.sleep(3)
    navegador.find_element(By.XPATH, '//*[@id="btn-save"]').click()
    time.sleep(1)
    i= i + 1

```
